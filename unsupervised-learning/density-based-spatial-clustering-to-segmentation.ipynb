# Using DBSCAN clustering to image segmentation

**Import library's**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Load Image**

```python
img = cv2.imread("/Users/neemiasbsilva/Downloads/Computer-Vision-with-Python/DATA/horse.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

**Visualizing the 3 channels in 3D plot**

```python
r, g, b = cv2.split(img)
print("\tShape of the Three Chaneels")
print("r: {}\t g: {}, b: {}".format(r.shape, g.shape, b.shape))
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot(r, g, b):
    fig = plt.figure(figsize=(12,12))
    
    ax = Axes3D(fig)
    ax.scatter(r, g, b)
    plt.show()
    
```

```python
r = r.flatten()
g = g.flatten()
b = b.flatten()

plot(r, g, b)
```
