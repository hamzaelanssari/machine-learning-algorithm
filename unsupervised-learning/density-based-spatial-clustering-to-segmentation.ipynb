## Apply DBSCAN Clustering

```python
vectorized = img.reshape((-1, 3))
```

```python
vectorized.shape
```

```python
 vectorized = np.float32(vectorized)
```

```python
from sklearn.cluster import DBSCAN
```

```python
clustering = DBSCAN(eps=0.5, min_samples=2, metric='euclidean', 
                    metric_params=None, algorithm='auto', 
                    leaf_size=30, p=None, n_jobs=None).fit(vectorized)
```

```python
clustering.labels_
```

```python
core_samples_mask = np.zeros_like(clustering.labels_, dtype=bool)
```

```python
core_samples_mask[clustering.core_sample_indices_] = True
```

```python
labels = clustering.labels_
```

**Number of cluster in labels, ignoring noise if present**

```python
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_clusters_
```

```python
n_noise_ = list(labels).count(-1)
n_noise_
```

```python
unique_labels = set(labels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(unique_labels))]
```

```python
len(colors)
```

```python
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (labels == k)

    xy = vectorized[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = vectorized[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```
