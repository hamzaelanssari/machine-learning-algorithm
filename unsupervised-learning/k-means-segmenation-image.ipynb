# Using K-means clustering to image segmentation

**Import library's**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Load Image**

```python
img = cv2.imread("/home/neemias/Imagens/dog.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

**Visualizing the 3 channels in 3D plot**

```python
r, g, b = cv2.split(img)
print("\tShape of the Three Chaneels")
print("r: {}\t g: {}, b: {}".format(r.shape, g.shape, b.shape))
```

```python
from mpl_toolkits.mplot3d import Axes3D


def plot(r, g, b):
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.scatter(r, g, b)
    plt.show()

    
r = r.flatten()
g = g.flatten()
b = b.flatten()

plot(r, g, b)
```

**Apply K-means Clustering**

```python
vectorized = img.reshape((-1,3))
```

```python
vectorized
```

```python
vectorized = np.float32(vectorized)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
```

```python
K = 2
attempts=10
ret,label,center=cv2.kmeans(vectorized,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
```

```python
center = np.uint8(center)
```

```python
res = center[label.flatten()]
result_image = res.reshape((img.shape))
```

```python
figure_size = 15
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,2,1),plt.imshow(img)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(result_image)
plt.title('Segmented Image when K = %i' % K), plt.xticks([]), plt.yticks([])
plt.show()
```
